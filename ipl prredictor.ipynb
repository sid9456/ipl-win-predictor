```python
import numpy as np
import pandas as pd
```

```python
deliveries=pd.read_csv('~/Downloads/deliveries.csv')
```

```python
matches=pd.read_csv('~/Downloads/matches.csv')
```

```python
deliveries.head()
```

```python
matches.head()
```

```python
# we need to have batting team,bowling team,city, runs_left,balls_left,wickets_left,total_runs crr,rrr, result
```

```python
total_score_df=deliveries.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total_score_df=total_score_df[(total_score_df['inning']==1)]
```

```python
total_score_df=total_score_df.reset_index()
```

```python
total_score_df=total_score_df.drop(['index'],axis=1)
```

```python
df=pd.concat([matches,total_score_df],axis=1)
```

```python
df=df.drop(['inning'],axis=1)
```

```python
df['team1'].unique()
```

```python
df['team1']=df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
df['team2']=df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

df['team1']=df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
df['team2']=df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
df['team1'].unique()
```

```python
teams=['Sunrisers Hyderabad', 'Mumbai Indians',
        'Royal Challengers Bangalore',
       'Kolkata Knight Riders', 'Delhi Capitals', 'Kings XI Punjab',
       'Chennai Super Kings', 'Rajasthan Royals',
       ]
```

```python
df=df[df['team1'].isin(teams)]
df=df[df['team2'].isin(teams)]
```

```python
df['team1'].unique()
```

```python
df.shape
```

```python
df['dl_applied'].value_counts()
```

```python
df=df[(df['dl_applied']==0)].reset_index()
```

```python
df=df.drop(['index'],axis=1)
```

```python
match_df=df[['match_id','city','winner','total_runs']]
```

```python
dely_df=match_df.merge(deliveries,on='match_id')
```

```python
deliveries.shape
```

```python
dely_df=dely_df[(dely_df['inning']==2)]
```

```python
dely_df.shape
```

```python
dely_df.head()
```

```python
dely_df['current_score']=dely_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
dely_df.head()
```

```python
dely_df['runs_left']=dely_df['total_runs_x']-dely_df['current_score']
```

```python
dely_df.head()
```

```python
dely_df['balls_left']=126-(dely_df['over']*6 + dely_df['ball'])
```

```python
dely_df.head()
```

```python
dely_df['player_dismissed']=dely_df['player_dismissed'].fillna('0')
dely_df['player_dismissed']=dely_df['player_dismissed'].apply(lambda x:x if x =='0' else 1)
dely_df['player_dismissed']=dely_df['player_dismissed'].astype('int')
```

```python
wickets=dely_df.groupby('match_id').cumsum()['player_dismissed'].values
dely_df['wickets']=10-wickets
```

```python
dely_df.head()
```

```python
dely_df['crr']=dely_df['current_score']*6/(120-dely_df['balls_left'])
```

```python
dely_df['rrr']=(dely_df['runs_left']*6)/dely_df['balls_left']
```

```python
dely_df.head()
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
dely_df['result']=dely_df.apply(result,axis=1)
```

```python
dely_df.head()
```

```python
final_df=dely_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df.head()
```

```python
final_df=final_df.sample(final_df.shape[0])
final_df.head()
```

```python
final_df.sample()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df=final_df[(final_df['balls_left']!= 0 )]
```

```python

from sklearn.model_selection import train_test_split
```

```python
X=final_df.drop(['result'],axis=1)
y=final_df['result']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf=ColumnTransformer([('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])]
                       ,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(steps=[('step1',trf),('step2',LogisticRegression(solver='liblinear'))])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
